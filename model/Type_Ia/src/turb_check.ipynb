```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 10
nx = 800
nz = 400
comp_list = []
for i in range(n):
    comp = np.array(np.loadtxt("/home/cnchong/Codes/cumc3d/model/Type_Ia/turb_comp"+str(i)+".dat"))
    comp = comp.reshape(nx,nz)
    comp_list.append(comp)

str_list = []
for i in range(n):
    stre = np.array(np.loadtxt("/home/cnchong/Codes/cumc3d/model/Type_Ia/turb_str"+str(i)+".dat"))
    stre = stre.reshape(nx,nz)
    str_list.append(stre)

eps_list = []
for i in range(n):
    eps = np.array(np.loadtxt("/home/cnchong/Codes/cumc3d/model/Type_Ia/turb_eps"+str(i)+".dat"))
    eps = eps.reshape(nx,nz)
    eps_list.append(eps)

RT_list = []
for i in range(n):
    RT = np.array(np.loadtxt("/home/cnchong/Codes/cumc3d/model/Type_Ia/turb_RT"+str(i)+".dat"))
    RT = RT.reshape(nx,nz)
    RT_list.append(RT)

F_list = []
for i in range(n):
    F = np.array(np.loadtxt("/home/cnchong/Codes/cumc3d/model/Type_Ia/turb_F"+str(i)+".dat"))
    F = F.reshape(nx,nz)
    F_list.append(F)

diff_list = []
for i in range(n):
    diff = np.array(np.loadtxt("/home/cnchong/Codes/cumc3d/model/Type_Ia/turb_diff"+str(i)+".dat"))
    diff = diff.reshape(nx,nz)
    diff_list.append(diff)
```
