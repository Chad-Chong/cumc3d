```python
import h5py
import numpy as np
import glob
import matplotlib.pyplot as plt
```

```python
files = glob.glob('./outfile/*.hdf5')
rhocgs2code = 1.619e-18
tcgs2code = 2.03017e5
rhocNM = []
rhocDM = []
timeCoor = []

for file in files:
    with h5py.File(file, 'r') as f:
        rhoc = (f['primitiveNM'][:].T[0][0,0])
        rhocNM.append(rhoc/rhocgs2code)
        rhoc = (f['primitiveDM'][:].T[0][0,0])
        rhocDM.append(rhoc/rhocgs2code)
        time = (f['time'][:][0])
        timeCoor.append(time/tcgs2code)

# Plot side by side
fig, ax = plt.subplots(2,1)
ax[0].plot(timeCoor, rhocNM)
ax[0].set_title('NM')
ax[1].plot(timeCoor, rhocDM)
ax[1].set_title('DM')
plt.xlabel('Time (s)')

plt.tight_layout()
plt.show()
```
